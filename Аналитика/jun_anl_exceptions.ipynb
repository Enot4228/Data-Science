### Вопросы по лекциям.

Как поймать вообще все ошибки, которые могут произойти?

**Ответ:**try:
           <Блок кода с ошибкой>
          except <Название ошибки>:
           <Обработка ошибки>


Сколько раз подряд можно указывать except?

**Ответ:** Столько сколько может быть ошибок поймать ошибок

**Ответ:**
try:
...
except <Название ошибки> as exc: print("...{exc}.....")

Вы хотите с помощью print вывести параметры ошибки в консоль, как это сделать?

**Ответ:**

Это предупреждение, говорящее о том, что в следующей версии может быть изменено то, что использовали сейчас

### Задание 1.

Есть файл с протоколом регистраций пользователей на сайте (registrations.txt).<br>
Каждая строка содержит информацию о имени, электронной почте и возрасте человека. <br><br>


Надо проверить данные из файла, для каждой строки:
 - присутсвуют все три поля
 - поле имени содержит только буквы
 - поле email содержит @ и .
 - поле возраст является числом от 10 до 99<br>

В результате проверки нужно сформировать два файла
 - registrations_good.log для правильных данных, записывать строки как есть
- registrations_bad.log для ошибочных, записывать строку и вид ошибки.<br>

Для валидации строки данных написать метод, который может выкидывать исключения:
 - НЕ присутсвуют все три поля: ValueError
 - поле имени содержит НЕ только буквы: NotNameError (кастомное исключение)
 - поле email НЕ содержит @ и .(точку): NotEmailError (кастомное исключение)
 - поле возраст НЕ является числом от 10 до 99: ValueError
Вызов метода обернуть в try-except.

```python
class NotNameError(Exception):
    pass


class NotEmailError(Exception):
    pass


st = '!'
with open('registrations_.txt', 'r') as reg_f:
    f_bad = open('registrations_bad.log', 'w')
    f_good = open('registrations_good.log', 'w')
    for line in reg_f:
        try:
            name, email, age = line.split(' ')
            print(line)
        except ValueError:
            print("Не достаточно данных")
            # Записать данные в bad
            f_bad.write(line)
            continue
        ################################################
        for i in name:
            try:
                if i == st: # Решил проверить только на один символ, так как у меня какие-то проблемы с кодировкой
                    raise NotNameError
                j = int(i)
                if type(j) == int:
                    num = j
                    num_2 = j
                    raise NotNameError
            except NotNameError:
                print("Не является именем")
                # Записать данные в bad
                f_bad.write(line)
                break
            except ValueError:
                continue
        try:
            if num == num_2:
                del num
                del num_2
                raise NotNameError
        except NotNameError:
            continue
        except NameError:
            print("Правильное имя")

        ####################################################
        for k in email:
            if k == '@':
                at = k
            elif k == '.':
                dot = k
        try:
            if at == '@' and dot == '.':
                print("Email правильный")
                del at
                del dot
            else:
                raise NotEmailError
        except NotEmailError:
            print("Email неправильный")
            # Записать данные в bad
            f_bad.write(line)
            continue
        except NameError:
            print("Email неправильный")
            # Записать данные в bad
            f_bad.write(line)
            continue
        #################################################
        try:
            age_int = int(age)
            if 10 < age_int < 99:
                print("Правильный возраст")
            else:
                raise ValueError
        except ValueError:
            print("Ошибки с возрастом")
            # Записать данные в bad
            f_bad.write(line)
            continue
        f_good.write(line)
    f_bad.close()
    f_good.close()
reg_f.close()
```
