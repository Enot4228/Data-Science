*Результат выполнения*

```python
file_path = './data/uwsgi.log'
ip_counter = dict()

with open(file_path, 'r', encoding='utf8') as f:
    for row in f:
        parted_row = row.split(' ')
        if len(parted_row) > 1:
            if parted_row[0] not in ip_counter:
                ip_counter[parted_row[0]] = 1
            else:
                ip_counter[parted_row[0]] += 1
            
        

print(ip_counter)
```

```python
import csv
columns_dict = dict()
file_path = './data/task.csv'


with open(file_path, newline='', encoding='utf8') as csvfile:
    reader = csv.DictReader(csvfile)
    pre_code = ''
    pre_idea = ''
    pre_component = ''
    pre_time = ''
    for col_name in reader.fieldnames:
        columns_dict[col_name] = []
    for row in reader:
        pre_code = pre_code + row['Код'] + ','
        code = pre_code.split(',')
        
        pre_idea = pre_idea + row['Тема'] + ','
        idea = pre_idea.split(',')
        
        pre_component = pre_component + row['Компонент'] + ','
        component = pre_component.split(',')
        
        pre_time = pre_time + row['Затрачено в часах'] + ','
        time = pre_time.split(',')
code.remove('')
idea.remove('')
component.remove('')
time.remove('')
columns_dict['Код'] = code
columns_dict['Тема'] = idea
columns_dict['Компонент'] = component
columns_dict['Затрачено в часах'] = time
print(columns_dict)
```

### 4.1
**Задание простого уровня** Распечатайте список курсов по программированию от Skillbox.

```python
import requests
from lxml import html

page = requests.get('https://skillbox.ru/code/').content

html_tree = html.fromstring(page)

items_list = html_tree.xpath("//b[@class='{}']".format('profession-card__title hover-card__text'))

for item in items_list:
    print(item.text)
```

### 4.2

**Задание среднего уровня**. Выведите список всех курсов Skillbox по программированию, дизайну, маркетингу и управлению. Добавьте к каждому курсу ссылку http на этот курс на сайте Skillbox.

```python
course_dict = {}
n = 0
links = html_tree.xpath("//div/a")
links = links[1:len(items_list)]
for link in links:
    if link.text not in course_dict:
        course_dict[items_list[n].text] = link.get('href')
    n += 1
course_dict
```

### 4.3
**Задание высокого уровня**. Напишите код, который переходит по ссылкам на страницы курсов по программированию из предыдущего задания и распечатывает описание курса.

```python
import requests

descr = []
links_href = []
for course in course_dict:
    
    page = requests.get(course_dict[course]).content

    html_tree = html.fromstring(page)

    items_list = html_tree.xpath("//p[@class='{}']".format('start-screen__desc'))
    for item in items_list:
        descr.append(item.text)      
    links_href.append(course_dict[course])
descr.insert(1, None)   
descr.insert(4, None)
descr.insert(10, None)
descr.insert(16, None)
result_df = pd.DataFrame({
    'link': links_href,
    'descr' : descr
})

result_df.head()
```
